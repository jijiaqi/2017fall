```python
def my_func(x):
    return -x[1]


a = [('a', 2), ('b', 3), ('c', 1)]
print ('before sort', a)
a.sort(key=my_func)
print ('after sort', a)

a = [('a', 2), ('b', 3), ('c', 1)]
a.sort(key=lambda zzz: -zzz[1])
print ('after sort', a)

def square(zzz):
    return zzz**2

a = [1, 2, 3, 4]
print map(square, a)
print map(lambda x:x**2, a)
```
